```python
import pandas as pd


ur1 = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data?authuser=0"
columns = ['id','diagnosis'] + [f'feature_{i}' for i in range(1,31)]
data = pd.read_csv(ur1, header=None, names=columns)

data.drop(columns=['id'] , inplace=True)

data.to_csv('breast_cancer_wisconsin.csv', index=False)

```

```python
data
```

```python
data.diagnosis.value_counts()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, accuracy_score
from imblearn.over_sampling import SMOTE
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
label_encoder = LabelEncoder()
data['diagnosis'] = label_encoder.fit_transform(data['diagnosis'])
```

```python
data.head()
```

```python
x = data.drop(columns=['diagnosis']).values
y = data['diagnosis'].values
```

```python
print(x, y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train,X_test
```

```python
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
```

```python
X_train_resampled.shape,  y_train_resampled.shape
```

```python
np.unique(y_train_resampled, return_counts = True)
```

```python
model = Sequential()
model.add(Dense(64, input_dim=X_train_resampled.shape[1], activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train_resampled, y_train_resampled, epochs=50, batch_size=32, validation_split=0.2)
```

```python
model.summary()
```

```python
y_pred = (model.predict(X_test) > 0.5).astype("int32")
print(classification_report(y_test, y_pred))
print("Accuracy:" ,accuracy_score(y_test, y_pred))
```
